```
df = pd.read_csv('us-jobs-by-age-summary.csv', thousands=',') # open csv file instruct pandas to interpret numbers with columns as thousands data will come in correctly typed
# create a new column for every age group that shows the percentage participation of age group in industry
df['16-19'] = (df['16 to 19'] / df['16 to 19'].sum()) * 100
df['20-24'] = (df['20 to 24'] / df['20 to 24'].sum()) * 100
df['25-34'] = (df['25 to 34'] / df['25 to 34'].sum()) * 100
df['35-44'] = (df['35 to 44'] / df['35 to 44'].sum()) * 100
df['45-54'] = (df['45 to 54'] / df['45 to 54'].sum()) * 100
df['55-64'] = (df['55 to 64'] / df['55 to 64'].sum()) * 100
df['65+ '] = (df['65+'] / df['65+'].sum()) * 100
df2 = df[['Industry', '16-19', '20-24', '25-34', '35-44', '45-54', '55-64', '65+ ']] # create new dataframe with just the new 'percentage participation' columns that we want to plot
df2 = df2.T # transpose the data frame
new_header = df2.iloc[0] #grab the first row for the header
df2 = df2[1:] #take the data less the header row
df2.columns = new_header #set the header row as the df header
df2.plot(figsize=(10, 15))
```
